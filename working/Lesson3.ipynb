# Ex 5: Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 8), sharex=True)

def func(x) :
    return x**3 + x**2 + x + 1

def derivative(f,x) :
    df = f[1:] - f[:-1]
    dx = x[1:] - x[:-1]
    return df/dx, x[:-1] + dx 
    

x = np.linspace(-10.,10,100000)

# Now you can access each axis via the axes array
axes[0].plot(x, func(x))
axes[0].set_title(r"$f(x) = x^3 + x^2 + x + 1$")
axes[0].set_ylabel(r"$f \ (x)$")

f1, x1 = derivative(func(x),x)

axes[1].plot(x1, f1)
axes[1].set_title("First derivative")
axes[1].set_ylabel(r"$f \ '(x)$")

f2, x2 = derivative(f1,x1)

axes[2].plot(x2, f2)
axes[2].set_title("Second derivative")
axes[2].set_ylabel(r"$f \ ''(x)$")
axes[2].set_xlabel(r"$x$")

plt.tight_layout()
plt.show()
```

# Ex 2 : Planet Positions

```python
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]
```

```python
plot = plt.figure(figsize=(7,7))
ax = plot.add_subplot(111)
ax.scatter(a,P)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Period (in Earth years)")
ax.set_ylabel("Distance from the Sun (AU)")

for i, label in enumerate(names):
    ax.annotate(label, (a[i], P[i]), xytext=(5, 7), ha='center', textcoords="offset points",)

plt.show()
```
